```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
""" Linear Scan """
linear_count = 0

def find_peak_linear(arr):
  global linear_count
  linear_count = 0
  n = len(arr)
  for i in range(n):
    linear_count += 1
    if (i == 0 or arr[i] >= arr[i - 1]) and (i == n - 1 or arr[i] >= arr[i + 1]):
      return arr[i]
  return None

```

```python
""" Divide and Conquer Method"""
def find_peak_divide_conquer(arr, low, high):
  global divide_conquer_count
  divide_conquer_count = 0

  def helper(arr, low, high):
    global divide_conquer_count
    mid = (low + high) // 2
    divide_conquer_count += 1
    n = len(arr)

    if (mid == 0 or arr[mid] >= arr[mid - 1]) and (mid == n - 1 or arr[mid] >= arr[mid + 1]):
      return arr[mid]
    elif mid > 0 and arr[mid] < arr[mid - 1]:
      return helper(arr, low, mid - 1)
    else:
      return helper(arr, mid + 1, high)

  return helper(arr, low, high)
```

```python
arr1 = [1, 3, 20, 4, 1]
arr2 = [1, 3, 5, 7, 9, 11, 13, 15]
arr3 = [10, 20, 15, 2, 23, 90, 67]

peak_linear_1 = find_peak_linear(arr1)
peak_divide_conquer_1 = find_peak_divide_conquer(arr1, 0, len(arr1) - 1)
print(f"Peak in arr1 (linear): {peak_linear_1}              (count: {linear_count})")
print(f"Peak in arr1 (divide and conquer): {peak_divide_conquer_1} (count: {divide_conquer_count})\n")

peak_linear_2 = find_peak_linear(arr2)
peak_divide_conquer_2 = find_peak_divide_conquer(arr2, 0, len(arr2) - 1)
print(f"Peak in arr2 (linear): {peak_linear_2}              (count: {linear_count})")
print(f"Peak in arr2 (divide and conquer): {peak_divide_conquer_2} (count: {divide_conquer_count})\n")

peak_linear_3 = find_peak_linear(arr3)
peak_divide_conquer_3 = find_peak_divide_conquer(arr3, 0, len(arr3) - 1)
print(f"Peak in arr3 (linear): {peak_linear_3}              (count: {linear_count})")
print(f"Peak in arr3 (divide and conquer): {peak_divide_conquer_3} (count: {divide_conquer_count})\n")


```
