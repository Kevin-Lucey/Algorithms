```python
import pandas as pd
data1 = {
    'ID': [1,2,3,4,5,6,7],
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Tom', 'Frank', "Frank"],
    'Age': [25, 30, 35, 40, 45, 50, 50],
}

df1 = pd.DataFrame(data1)
print(df1)
```

```python
df_no_duplicates = df1.drop_duplicates(subset=['Name', 'Age'])
print("df1: ")
print(df1)
print("df_no_duplicates: ")
print(df_no_duplicates)
```

```python
data2 = {
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Tom', 'Frank'],
    'Age': [25, 30, 35, None, 45, 50],
    'Salary': [50000, 60000, 70000, 80000, None , 35000]
}
df2 = pd.DataFrame(data2)
df2
```

```python
# Deletion

df2_no_missing = df2.dropna()
print(df2_no_missing.shape)
df2
```

```python
df2_deleted_columns = df2.dropna(axis=1)
print(df2_deleted_columns.shape)
df2_deleted_columns
```

```python
df3 = df2.fillna(0)
print(df3)
```

```python
mean_age = df2['Age'].mean()
mean_salary = df2['Salary'].mean()

df_mean = df2.copy()

df_mean['Age'].fillna(mean_age, inplace=True)
df_mean['Salary'].fillna(mean_salary, inplace=True)

print(df_mean)
```

```python
median_age = df2['Age'].median()
median_salary = df2['Salary'].median()

df_median = df2.copy()

df_median['Age'].fillna(median_age, inplace=True)
df_median['Salary'].fillna(median_salary, inplace=True)

print(df2)
```

```python
df2_mean = df2.copy()

df2_mean = df2_mean.assign(
    Age= df2_mean['Age'].fillna(df2_mean['Age'].mean()),
    Salary = df2_mean['Salary'].fillna(df2_mean['Salary'].mean())
)

df2_mean
```

```python
df2_median = df2.copy()

df2_median = df2_median.assign(
    Age= df2_median['Age'].fillna(df2_median['Age'].median()),
    Salary = df2_median['Salary'].fillna(df2_median['Salary'].median())
)

df2_median
```

```python
df2_mode = df2.copy()

df2_mode = df2_mode.assign(
    Age= df2_mode['Age'].fillna(df2_mode['Age'].mode()),
    Salary = df2_mode['Salary'].fillna(df2_mode['Salary'].mode())
)

df2_mode
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

data = np.array([
    [10, 200],
    [20,250],
    [30, 300],
    [40,350],
    [50, 400],
    [60, 450],
    [70, 500],
    [80, 550],
    [90, 600],
    [100, 650]
])

df = pd.DataFrame(data, columns=['Feature1', 'Feature2'])
print(df)
```

```python
scaler_minmax = MinMaxScaler()

normalized_data = scaler_minmax.fit_transform(df)
normalized_data = pd.DataFrame(normalized_data, columns=df.columns)

normalized_data.head()
normalized_data.describe()
```

```python
scaler_standard = StandardScaler()

standardized_data = scaler_standard.fit_transform(df)

df_standardized = pd.DataFrame(standardized_data, columns = df.columns)

df_standardized
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([10,12,11,12,14,15,18,100])

fig = plt.figure(figsize=(10,7))
plt.boxplot(data)
plt.title('Outlier Detection')
plt.grid(True)
plt.show()
```

```python
data = np.array([10,12,11,12,14,15,18,10])

fig = plt.figure(figsize=(10,7))
plt.boxplot(data)
plt.title('Outlier Detection')
plt.grid(True)
plt.show()
```

```python
from scipy import stats

data = np.array([10,12,12,13,14,15,16,17,18,100])

z_scores = stats.zscore(data)

threshold = 2.5

outliers = np.where(np.abs(z_scores)>threshold)

print("Z-scores: ", z_scores)
print("Outliers: ", outliers)
print("Outlier values: ", data[outliers])
```

```python
from sklearn.preprocessing import LabelEncoder

data = ['Red', 'Green', 'Blue', 'Red', 'Green']

label_encoder = LabelEncoder()

encoded_data = label_encoder.fit_transform(data)

print('Encoded Data: ', encoded_data)
```
