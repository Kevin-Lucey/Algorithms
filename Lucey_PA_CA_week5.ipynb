```python
import random
import time
```

```python
"""Bogo Sort"""

def bogoSort(a):
  n = len(a)
  while not isSorted(a):
    shuffle(a)
    print("Current array state: ", a)

def isSorted(a):
  n = len(a)
  for i in range(0, n-1):
    if a[i] > a[i+1]:
      return False
  return True

def shuffle(a):
  n = len(a)
  for i in range(0, n):
    r = random.randint(0, n-1)
    a[i], a[r] = a[r], a[i]

a = [3,2,4,0]
bogoSort(a)
print("===========")
print("Sorted array: ")
print(" ".join(str(x) for x in a))
print("===========")

```

```python
"""Bubble Sort"""

def bubbleSort(arr):
  n = len(arr)

  for i in range(n):
    for j in range(0, n-i-1):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
      print("Step: ", arr)
    print("After pass ", i+1, ":", arr)
    print("")
  return arr
arr = [64, 34, 25, 12, 22, 11, 90]
print("Original array: ", arr)
sorted_arr = bubbleSort(arr)
print("Sorted array: ", sorted_arr)
```

```python
"""Selection Sort"""

def selectionSort(arr):
  n = len(arr)
  for i in range(n):
    min_idx = i
    for j in range(i+1, n):
      if arr[j] < arr[min_idx]:
        min_idx = j

    arr[i], arr[min_idx] = arr[min_idx], arr[i]
    print("After iteration ", i+1, ": ", arr)
  return arr

arr = [64, 25, 12, 22, 11]
print("Original array: ", arr)
sorted_arr = selectionSort(arr)
print("Sorted array: ", sorted_arr)
```

```python
"""New testing arrays"""

test_arr = [43, 23, 5, 10, -3, 18, 100, 4, 13, 57, -6, 120, 56, 78, 67]

print("Original array: ", test_arr)
print("===============")
print("Selection Sort")

start = time.perf_counter()
sorted_arr = selectionSort(test_arr)
end = time.perf_counter()

print("Sorted array: ", sorted_arr)
print("Time: ", end-start)
print("================")
print("Bubble Sort")

start = time.perf_counter()
sorted_arr = bubbleSort(test_arr)
end = time.perf_counter()

print("Sorted array: ", sorted_arr)
print("Time: ", end-start)
print("================")
print("\n Bogo Sort")

start1 = time.perf_counter()
bogoSort(test_arr)
end1 = time.perf_counter()

print("Sorted array: ", test_arr)
print("Time: ", end1-start1)
```

The algorithm that was fastest was Bogo sort in some instances. On some executions it would be faster due to the probability of a list randomly being sorted in order. Overall, Selection sort was the fastest with a consistent time of 0.0002 seconds while Bubble sort was consitently 0.0009 seconds.

Bubble sort seemed to be the slowest for this sized array at 0.0009 seconds while the slowest runtime for Bogo sort was 0.0006 in my observered runtimes.

Even though Bogo showed good results, this algorithm is impractical at scale when you try nad sort a list of 100s, 1000s, or 100,000s of values whereas Bubble and Selction Sort offer consistent runtimes.
